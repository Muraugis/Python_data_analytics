Processing the data

```python
# Removing all buildings built pre 1900 or with missing build date

df = df[df["build_year"] >= 1900] 

# Adding a new column to bucket into 10-year build ranges

buckets = range(1900, 2030, 10)

df['build_year_range'] = pd.cut(df['build_year'], bins=buckets)

# Adding another column to indicate period during which apartment was built

def period_value(build_year):
    if build_year < 1914:
        return 'Pre-WW1'
    if build_year < 1918:
        return 'During WW1'
    elif build_year < 1940:
        return 'Interwar Independence'
    elif build_year < 1945:
        return 'During WW2'
    elif build_year < 1991:
        return 'Soviet Occupation'
    else:
        return 'Modern Independence'

df['period'] = df['build_year'].apply(lambda x: period_value(x))
```

Selecting the required columns and saving file to CSV
